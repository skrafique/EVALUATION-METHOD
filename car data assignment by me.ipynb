# Logistic Regression

```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the data into dataframe
df = pd.read_csv("car data.csv")
```

```python
df.head()
```

```python
plt.scatter(df["Selling_Price"],df["Owner"])
```

```python
plt.scatter(df["Present_Price"],df["Owner"],color = "r")
```

```python
plt.scatter(df["Kms_Driven"],df["Owner"],color = "g")
```

```python
# partitioning the x and y value
x = df.iloc[:,[2,3,4]]
y = df.iloc[:,-1]
```

```python
x
```

```python
#importing train_test_split to train and test the data
from sklearn.model_selection import train_test_split
```

```python
# performing  the train and test data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 4)
```

```python
# importing Logistic Regression model
from sklearn.linear_model import LogisticRegression
```

```python
# creating the object
model = LogisticRegression()
```

```python
# fitting the model
model.fit(x_train,y_train)
```

```python
# predicting the x_test data
y_pred = model.predict(x_test)
```

```python
y_test
```

```python
# for comparing betweeen the y_test and y_pred
from sklearn.metrics import confusion_matrix
c_mat = confusion_matrix(y_pred,y_test)
```

```python
c_mat
```

```python
# test the score
model.score(x_test,y_test)
```

### for Logistic Regression accuracy_score of x_test and y_test is 94%

# Decision Tree classifier

```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the data into dataframe
df = pd.read_csv("car data.csv")
```

```python
df.head()
```

```python
# partitioning the x and y value
x = df.iloc[:,[2,3,4]]
y = df.iloc[:,-1]
```

```python
#importing train_test_split to train and test the data
from sklearn.model_selection import train_test_split
```

```python
# performing  the train and test data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 4)
```

```python
# creating object
model1 = DecisionTreeClassifier()
```

```python
# predicting the x_test data
y_pred = model1.predict(x_test)
```

```python
y_test
```

```python
# test the score
model1.score(x_test,y_test)
```

### for DecisionTreeClassifier  accuracy score for x_test and y_test is 91%
